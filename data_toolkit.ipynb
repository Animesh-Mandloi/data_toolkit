```python
"""generate an array of evenly spaced numbers between two values and then reshape it into a 2D array, you can use numpy.linspace and numpy.reshape"""
import numpy as np


start = 1 
end = 10   
num_elements = 12  


array_1d = np.linspace(start, end, num_elements)


array_2d = array_1d.reshape((3, 4))  

print("1D Array:")
print(array_1d)
print("\n2D Array:")
print(array_2d)
```

```python
#Creating a Random Integer Array and Extracting Even and Odd Numbers

import numpy as np


array_2d = np.random.randint(1, 11, size=(5, 5))

even_numbers = array_2d[array_2d % 2 == 0]

odd_numbers = array_2d[array_2d % 2 != 0]

print("2D Array:")
print(array_2d)
print("\nEven Numbers:")
print(even_numbers)
print("\nOdd Numbers:")
print(odd_numbers)
```

```python
#Working with a 3D NumPy Array

import numpy as np


array_3d = np.random.randint(1, 11, size=(4, 4, 4))


max_indices = np.argmax(array_3d, axis=2)


array_transpose = array_3d.transpose((1, 0, 2))
elementwise_multiplication = array_3d * array_transpose[:4, :4, :]

print("3D Array:")
print(array_3d)
print("\nIndices of Maximum Values (along the third axis):")
print(max_indices)
print("\nElement-wise Multiplication Result:")
print(elementwise_multiplication)
```

```python
"""Create a 3D NumPy array of shape (4, 4, 4) containing random integers between 1 and 10, find the indices of the maximum values along each depth level (third axis), and perform element-wise multiplication between the array and its transpose."""

import numpy as np


array_3d = np.random.randint(1, 11, size=(4, 4, 4))


max_indices = np.argmax(array_3d, axis=2)


array_transpose = array_3d.transpose(1, 0, 2)  
elementwise_multiplication = array_3d * array_transpose[:4, :4, :]

print("3D Array:")
print(array_3d)
print("\nIndices of Maximum Values (along the third axis):")
print(max_indices)
print("\nElement-wise Multiplication Result:")
print(elementwise_multiplication)
```

```python
#Plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine Wave Function'.

from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

# Prepare output to display in notebook
output_notebook()

# Generate data for the sine wave
x = np.linspace(0, 4 * np.pi, 100)
y = np.sin(x)

# Create a new figure
p = figure(title="Sine Wave Function", x_axis_label='x', y_axis_label='sin(x)')

# Add a line renderer
p.line(x, y, line_width=2)

# Add grid lines
p.grid.grid_line_alpha = 0.3

# Show the plot
show(p)
```

```python
#Generate a bar chart of randomly generated categorical data, color bars based on their values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical Bar Chart'.

from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource, HoverTool
import numpy as np
import pandas as pd

# Prepare output to display in notebook
output_notebook()

# Generate random categorical data
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = np.random.randint(10, 100, size=len(categories))

# Create a data source
source = ColumnDataSource(data=dict(categories=categories, values=values))

# Create a new figure
p = figure(x_range=categories, title="Random Categorical Bar Chart", x_axis_label='Category', y_axis_label='Value', plot_height=350)

# Add a bar renderer
bars = p.vbar(x='categories', top='values', width=0.9, source=source, legend_field="categories", line_color='white', fill_color='color')

# Add hover tooltips
hover = HoverTool()
hover.tooltips = [("Category", "@categories"), ("Value", "@values")]
p.add_tools(hover)

# Show the plot
show(p)
```
